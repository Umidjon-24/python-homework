```python
import pandas as pd
df_json = pd.read_json("iris.json")
df_json
```

```python
sepalLength_mean = df_json['sepalLength'].mean()
sepalWidth_mean = df_json['sepalWidth'].mean()
petalLength_mean = df_json['petalLength'].mean()
petalWidth_mean = df_json['petalWidth'].mean()
print(sepalLength_mean, sepalWidth_mean, petalLength_mean, petalWidth_mean)

sepalLength_median = df_json['sepalLength'].median()
sepalWidth_median = df_json['sepalWidth'].median()
petalLength_median = df_json['petalLength'].median()
petalWidth_median = df_json['petalWidth'].median()
print(sepalLength_median, sepalWidth_median, petalLength_median, petalWidth_median)

sepalLength_std = df_json['sepalLength'].std()
sepalWidth_std = df_json['sepalWidth'].std()
petalLength_std = df_json['petalLength'].std()
petalWidth_std = df_json['petalWidth'].std()
print(sepalLength_std, sepalWidth_std, petalLength_std, petalWidth_std)
```

```python
df_titanic = pd.read_excel("titanic.xlsx")
minimum_age = df_titanic['Age'].min()
maximum_age = df_titanic['Age'].max()
sum_age = df_titanic['Age'].sum()
print(minimum_age, maximum_age, sum_age)
```
