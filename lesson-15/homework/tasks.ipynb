Task-1 Basic Plotting

```python
import numpy as np
import math
from matplotlib import pyplot as plt

x = np.arange(-10,10)
y = x**2-4*x+4
plt.plot(x,y)
plt.show()
```

Task-2 Sine and Cosine plot

```python
x = np.arange(0,2*math.pi, 0.1)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y, 'r--', label='sin(x)')
plt.plot(x,z, 'g', label='cos(x)')
plt.xlabel('angles in radians')
plt.ylabel('function y')
plt.legend()
plt.show()
```

Task-3 Subplots

```python
plt.subplot(2,2,1)
y = x**3
plt.plot(x,y, 'r--')
plt.title("y=x^3 graph")
plt.xlabel('x values')
plt.ylabel('x^3 values')

plt.subplot(2,2,2)
j = np.arange(0,2*math.pi)
z = np.sin(j)
plt.plot(j,z, 'g-.')
plt.title("y=sin(x) graph")
plt.xlabel('x values')
plt.ylabel('sin(x) values')

plt.subplot(2,2,3)
b = np.arange(-5,5)
a = math.e**b
plt.plot(b,a, 'b')
plt.title("y=e^x graph")
plt.xlabel('x values')
plt.ylabel('e^x values')

plt.subplot(2,2,4)
c = np.arange(0,10)
d = np.log10(c+1)
plt.plot(c,d, 'k:')
plt.title("y=log10(x) graph")
plt.xlabel('x values')
plt.ylabel('log10(x) values')

plt.subplots_adjust(wspace=0.5, hspace=0.5)
plt.show()
```

Task-4 Scatter Plot

```python
x = np.random.uniform(0,10,100)
y = np.random.uniform(0,10,100)
sizes = np.random.randint(0,50,100)
plt.scatter(x,y, sizes=sizes, color='blue')
plt.colorbar()
plt.xlabel('x values')
plt.ylabel('y values')
plt.title("Uniform scatter plot")
plt.show()
```

Task-5 Histogram

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, color='skyblue', alpha=0.5, edgecolor='red')
plt.title("Normal distribution histogram")
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

Task-6 3D Plotting

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)

xx, yy = np.meshgrid(x, y)

zz = np.cos(xx**2 + yy**2)

ax.set_xlabel("x values")
ax.set_ylabel("y values")
ax.set_zlabel("z values")
ax.plot_surface(xx, yy, zz, cmap='plasma')
```

Task-7 Bar Chart

```python
x = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
y = [200, 150, 250, 175, 225]
colors = ['red', 'blue', 'yellow', 'green', 'black']
plt.bar(x,y, color=colors)
plt.title("Sales data for 5 different products")
plt.xlabel("Products")
plt.ylabel("Sales values")
plt.show()
```

Task-8 Stacked Bar Chart

```python
time_periods = ['T1', 'T2', 'T3', 'T4']
category_A = [3, 5, 7, 6]
category_B = [2, 3, 4, 3]
category_C = [4, 2, 5, 7]

bar_width = 0.6

# Stacking the bars
plt.bar(time_periods, category_A, label='Category A', color='b')
plt.bar(time_periods, category_B, bottom=category_A, label='Category B', color='r')
plt.bar(time_periods, category_C, bottom=np.array(category_A) + np.array(category_B), label='Category C', color='g')

plt.title('Stacked Bar Chart of Categories Over Time')
plt.xlabel('Time Periods')
plt.ylabel('Values')
plt.legend()
plt.show()
```
